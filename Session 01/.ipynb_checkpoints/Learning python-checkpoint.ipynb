For Our Scope You don't need to know a lot about classes and Inheritence but we will cover it shortly

```python
class Father:
    def __init__(self):
        self.a = 20
```

```python
class Son(Father):
    def __init__(self):
        Father.__init__(self)
        self.b = 40
```

```python
s = Son()
```

```python
s.b
```

```python
s.a
```

```python
issubclass(Son, Father)
```
